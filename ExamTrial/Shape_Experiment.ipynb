```python
prova = np.zeros((18,10,1))
for i in range(18):
  print(i)
prova[:,:,0] = np.arange(0,10)
#prova
```

```python
def normalize_series(data, min, max):
    data = data - min
    data = data / max
    return data
```

```python
def make_fake_windows(BATCH_SIZE,N_PAST,N_FUTURE):
  tot_series = np.random.randint(low = 1, high = 10, size=BATCH_SIZE * (N_PAST+N_FUTURE))
  tot_series = normalize_series(tot_series, np.amin(tot_series), np.amax(tot_series))
  window_step = np.expand_dims(np.arange(N_PAST+N_FUTURE), axis=0)
  window_indexes = window_step + np.expand_dims(np.arange(len(tot_series)-(N_PAST+N_FUTURE-1)), axis=0).T # create 2D array of windows of size window_size
  tot_series = tot_series[window_indexes]
  features = tot_series[:,:N_PAST]
  labels = tot_series[:,N_PAST:]
  return (np.expand_dims(features, axis=-1),np.expand_dims(labels, axis=-1))
  
```

```python
(features,labels)  = make_fake_windows(50,10,10)
print(labels.shape)
#print(labels)
print(features.shape)
#print(features)
```
